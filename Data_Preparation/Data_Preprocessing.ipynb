### Adding All DFs Together: News Sentiments, Glassnode Data, Indices, Commodoties, USA Infilation and Interest Rates

```python
# Read the CSV file
daily_news_sentiments_df     = pd.read_csv('daily_news_sentiments.csv')
glassnode_data_df            = pd.read_csv('glassnode_data.csv')
indices_rates_commodoties_df = pd.read_csv('indices_rates_commodoties.csv')

# weather there are any NaN values in the dfs
# print(daily_news_sentiments_df.isna().sum())
# print(glassnode_data_df.isna().sum())
# print(indices_rates_commodoties_df.isna().sum())

merged_df = daily_news_sentiments_df.merge(glassnode_data_df, on='date', how='outer')
merged_df = merged_df.fillna(0) # because no sentiments previous so assigning 0
# Sort the DataFrame by the 'Date' column
merged_df = merged_df.sort_values('date')
# Reset the index
merged_df = merged_df.reset_index(drop=True)
merged_df = merged_df.iloc[:4136]


merged_df = merged_df.merge(indices_rates_commodoties_df, on='date', how="outer")
merged_df = merged_df.sort_values('date')
# Reset the index
merged_df = merged_df.reset_index(drop=True)

# forward and backward filling the NaN values which correspond with the weekends or backholidays
merged_df = merged_df.fillna(method='ffill')
merged_df = merged_df.fillna(method='bfill')

merged_df
```

### RSI and Other indicators

```python
# !pip install ta
```

```python
import pandas as pd
from ta.trend import SMAIndicator, MACD
from ta.momentum import RSIIndicator

# Moving Average (MA) with a window of 14
sma_indicator = SMAIndicator(merged_df['close'], window=14)
merged_df['SMA_14'] = sma_indicator.sma_indicator()

# Relative Strength Index (RSI)
rsi_indicator = RSIIndicator(merged_df['close'])
merged_df['RSI'] = rsi_indicator.rsi()

# Moving Average Convergence Divergence (MACD)
macd_indicator = MACD(merged_df['close'])
merged_df['MACD'] = macd_indicator.macd()

merged_df = merged_df.fillna(method='bfill')
print(merged_df.head(30))
```

### Year, Time Processing

```python
# Converting 'date' to datetime if it's not already
import numpy as np
# df = pd.DataFrame()
merged_df['date'] = pd.to_datetime(merged_df['date'])

# Extracting day and month from date and perform 2D transformation using sin and cos
merged_df['day_sin'] = np.sin(2 * np.pi * merged_df['date'].dt.day/merged_df['date'].dt.days_in_month)
merged_df['day_cos'] = np.cos(2 * np.pi * merged_df['date'].dt.day/merged_df['date'].dt.days_in_month)
merged_df['month_sin'] = np.sin(2 * np.pi * merged_df['date'].dt.month/12)
merged_df['month_cos'] = np.cos(2 * np.pi * merged_df['date'].dt.month/12)

# Extracting year from the date
merged_df['year'] = merged_df['date'].dt.year

merged_df = merged_df.drop('date', axis=1)

print(merged_df.head(30))
```

```python
merged_df.to_csv('Final_data_daily.csv', index=False)
```

```python
from sklearn.preprocessing import MinMaxScaler


scaler = MinMaxScaler()

cols_to_exclude = ['day_sin', 'day_cos', 'month_sin', 'month_cos']

# applying scaler to all columns in df except the excluded ones
merged_df_scaled = pd.DataFrame(scaler.fit_transform(merged_df.drop(cols_to_exclude, axis=1)), 
                         columns=merged_df.drop(cols_to_exclude, axis=1).columns)

# copy the excluded columns to the new scaled dataframe
for col in cols_to_exclude:
    merged_df_scaled[col] = merged_df[col]
```

```python
merged_df_scaled.to_csv('Final_data_scaled_daily.csv', index=False)
```
