```python
import pandas as pd
import requests
import json

dfs = {}

for key in requests_sorted_dict:
    # getting request details
    request_details = requests_dict[key]
    table_name = request_details['table_name']
    url = request_details['url']
    params = request_details['params']

    # making API request
    res = requests.get(url, params=params)

    # checking if request was successful
    if res.status_code == 200:
        
        # loading response into a DataFrame
        data = json.loads(res.text)
        temp_df = pd.DataFrame(data)
        temp_df['t'] = pd.to_datetime(temp_df['t'], unit='s')

        dfs[table_name] = temp_df
    else:
        print(f"Request for key {key} failed with status code {res.status_code}")
```

```python
def concat_dfs_on_time(dfs_dict, time_col, new_col_names):
    """
    Concatenates multiple dataframes on a specified time column and assigns new column names.

    Parameters:
    - dfs_dict: A dictionary where keys are table name and values are the dataframes to be concatenated.
    - time_col: A string indicating the column name in each dataframe that represents the time. This column will be used as the index for the merged dataframe.
    - new_col_names: A list of strings where each string is a new column name for the merged dataframe. The length of this list should match the number of dataframes in dfs_dict.

    Returns:
    - A single dataframe resulting from the concatenation of all dataframes in dfs_dict, indexed by time_col and with columns renamed as per new_col_names.
    """

    dfs = list(dfs_dict.values())
    for df, new_col_name in zip(dfs, new_col_names):
        # setting time col as index
        df.set_index(time_col, inplace=True)

        # renaming the columns you want to merge
        df.columns = [new_col_name]

    # concatinating dataframes
    result_df = pd.concat(dfs, axis=1)

    # filling missing values with nan
    result_df.fillna(value='NaN', inplace=True)
    
    return result_df
```

```python
import pandas as pd

def normalise_dict_column(df, column):
    """
    Takes a DataFrame and a column name containing dictionary values.
    Normalises the dictionary values into a flat table structure.
    
    Parameters:
    - df (DataFrame): The input DataFrame.
    - column (str): The column name containing dictionary values.
    
    Returns:
    - DataFrame: A new DataFrame with the normalized values.
    """
    temp_df = pd.json_normalize(df[column])
    return temp_df

def complete_dict_column_naming_conv(df, extra=""):
    """
    Renames the columns of a DataFrame by prefixing them with a given string.
    
    Parameters:
    - df (DataFrame): The input DataFrame.
    - extra (str): The string to prefix to each column name. Default is an empty string.
    
    Returns:
    - DataFrame: The DataFrame with renamed columns.
    """
    df.columns = extra + " " + df.columns
    return df

def normalise_cols(col_name):
    """
    Normalises a column containing dictionary values from a global DataFrame 'dfs'.
    Also renames the resulting columns by prefixing them with the column name.
    
    Parameters:
    - col_name (str): The column name in 'dfs' containing dictionary values.
    
    Returns:
    - DataFrame: A new DataFrame with normalised and renamed columns.
    """
    normalized_df = normalise_dict_column(dfs[col_name], col_name)
    normalized_df = complete_dict_column_naming_conv(normalized_df, extra=col_name)
    
    return normalized_df
```

```python
# HODL Waves Col
HODL_Waves_DF = normalise_cols('HODL Waves')
# start date
start_date = '2011-12-31'
# generating date range so that the new created df can be contatinated with new_df dataframe
dates = pd.date_range(start_date, periods=4138)

# setting new index as dates
HODL_Waves_DF.set_index(dates, inplace=True)

# Realized Cap HODL Waves Col
Realized_Cap_HODL_Waves_DF = normalise_cols('Realized Cap HODL Waves')

# setting new index as dates
Realized_Cap_HODL_Waves_DF.set_index(dates, inplace=True)

# Stock-to-Flow Ratio Col
Stock_to_Flow_Ratio_DF = normalise_cols('Stock-to-Flow Ratio')

# setting new index as dates
Stock_to_Flow_Ratio_DF.set_index(dates, inplace=True)
```
