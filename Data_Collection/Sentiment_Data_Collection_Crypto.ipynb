```python
import requests
from pprint import pprint as pp
import pickle

# https://cryptonews-api.com/api/v1/stat?&tickers=BTC,ETH&date=01152019-today&page=1&token=zd3vl8bkcmu0tkbmcsf1am2dpr12hfbyhcoetidy

def get_all_pages():

    """
    API Doc Link: https://cryptonews-api.com/documentation
    
    Retrieves all cryptocurrency news sentiment statistics for Bitcoin and Ethereum from CryptoNews API.
    
    Returns:
    - List of JSON responses containing the cryptocurrency news statistics. Each entry in the list corresponds to a page of data.
    
    Notes:
    - Please ensure that you insert your own API token.
    - The function prints "Something went wrong" if the request encounters an error.
    """

    base_url = 'https://cryptonews-api.com/api/v1/stat'
    params = {
        'tickers': 'BTC,ETH',
        'date': '01152019-today',
        'token': 'zd3vl8bkcmu0tkbmcsf1am2dpr12hfbyhcoetidy',
    }

    page = 1
    responses = []
    while True:
        params['page'] = page
        response = requests.get(base_url, params=params)
        
        if response.status_code == 200:
            data = response.json()
            if not data:
                break
            responses.append(data)
            page += 1
            
        else:
            print("Something went wrong")
            break

    return responses


responses = get_all_pages()
for response in responses:
    pp(response)

# pickling the list
with open('raw_news_sentiment_BTC_ETH.pkl', 'wb') as f:
    pickle.dump(responses, f)
```
